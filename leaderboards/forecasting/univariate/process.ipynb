```python
df_freq = df.groupby('freq').mean()
df_freq = df_freq.drop(['Unnamed: 0','Trend',	'Seasonality',	'Stationarity',	'Transition',	'Shifting'], axis=1)
df_freq
```

```python
metrics = ['mae','mape','mase','mse','msmape','rmse','smape','wape']
for metric in metrics:
    df = pd.read_csv(f'./result/{metric}.csv')
    dataset_list=['Seasonality','freq','Trend', 'Stationarity',	'Transition',	'Shifting']
    dfs=[]
    for col in dataset_list:
        temp_list = dataset_list.copy()
        temp_list.remove(col)
        if col != 'freq':
            temp_list.remove('freq')
        df_temp = df.groupby(col).mean()
        df_temp = df_temp.drop(temp_list + ['Unnamed: 0'], axis=1)
        df_temp = df_temp.rename(index=lambda val: f"{col}_{val}").rename_axis("Dataset").reset_index()
        dfs.append(df_temp)
    df_cat = pd.concat(dfs)
    df_cat.to_csv(f'./result/processed/{metric}.csv',index=False)
```

```python
import pandas as pd
metrics = ['mae','mape','mase','mse','msmape','rmse','smape','wape']
dfs=[]
for metric in metrics:
    df = pd.read_csv(f'./result/processed/{metric}.csv')
    dfs.append(df)
df = pd.concat(dfs,axis=0)
df.to_csv(f'./result/processed/all.csv')
```
